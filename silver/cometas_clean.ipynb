```python
%sql
-- Tabla en silver para los cometas
USE CATALOG nasa_dw;

CREATE OR REPLACE TABLE silver.cometas_clean
USING DELTA
COMMENT 'Datos de cometas limpios y normalizados'
AS
SELECT
    designacion                                         AS cometa_id,
    designacion                                         AS nombre,
    TO_DATE(fecha_acercamiento, 'yyyy-MMM-dd HH:mm')   AS fecha_acercamiento,
    ROUND(distancia_au, 6)                              AS distancia_au,
    ROUND(distancia_km, 2)                              AS distancia_km,
    ROUND(distancia_km / 384400, 4)                     AS distancia_lu,
    ROUND(velocidad_kms, 4)                             AS velocidad_kms,
    ROUND(velocidad_kmh, 2)                             AS velocidad_kmh,
    magnitud_absoluta,
    tipo_astro,
    CASE
        WHEN designacion LIKE 'P/%'
          OR designacion RLIKE '^[0-9]+P'              THEN 'Corto'
        WHEN designacion LIKE 'C/%'                    THEN 'Largo'
        ELSE 'Indeterminado'
    END                                                 AS periodo,
    CURRENT_TIMESTAMP()                                 AS fecha_procesamiento
FROM nasa_dw.bronze.cometas_raw;
```
